```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Criar um grafo
G = nx.Graph()

# Adicionar nós e arestas com pesos
edges = [
    ('A', 'B', 1),
    ('B', 'C', 2),
    ('A', 'C', 4),
    ('C', 'D', 1),
    ('B', 'D', 5)
]

for u, v, w in edges:
    G.add_edge(u, v, weight=w)

# Calcular o caminho mais curto usando Dijkstra
source = 'A'
target = 'D'
shortest_path = nx.dijkstra_path(G, source, target)

# Visualizar o grafo
pos = nx.spring_layout(G)  # Layout do grafo

# Desenhar nós e arestas
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_weight='bold')

# Desenhar as arestas com os pesos
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10)

# Destacar o caminho mais curto
path_edges = list(zip(shortest_path, shortest_path[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='r', width=2)

# Mostrar o grafo
plt.title(f"Caminho mais curto de {source} a {target}: {' -> '.join(shortest_path)}")
plt.show()
```

# Exemplo de Redes de Transporte e Logística

- Criar distâncias a partir de localização
- Precisa estar em formato array ou serve em matriz?


Vamos considerar as seguintes cidades conectadas por estradas com as distâncias indicadas:

Cidade A a Cidade B: 5 km
Cidade A a Cidade C: 10 km
Cidade B a Cidade C: 2 km
Cidade B a Cidade D: 4 km
Cidade C a Cidade D: 7 km
Cidade C a Cidade E: 1 km
Cidade D a Cidade E: 3 km

```python
# Criar um grafo
G = nx.Graph()

# Adicionar nós e arestas com pesos (distâncias)
edges = [
    ('A', 'B', 5),
    ('A', 'C', 10),
    ('B', 'C', 2),
    ('B', 'D', 4),
    ('C', 'D', 7),
    ('C', 'E', 1),
    ('D', 'E', 3)
]

for u, v, w in edges:
    G.add_edge(u, v, weight=w)

# Definir cidades de origem e destino
source = 'A'
target = 'E'

# Calcular o caminho mais curto usando Dijkstra
shortest_path = nx.dijkstra_path(G, source, target)

# Visualizar o grafo
pos = nx.spring_layout(G)  # Layout do grafo

# Desenhar nós e arestas
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10, font_weight='bold')

# Desenhar as arestas com os pesos
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10)

# Destacar o caminho mais curto
path_edges = list(zip(shortest_path, shortest_path[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='r', width=2)

# Mostrar o grafo
plt.title(f"Caminho mais curto de {source} a {target}: {' -> '.join(shortest_path)}")
plt.show()
```
