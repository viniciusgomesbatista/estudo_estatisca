# Correlação de Peason

* Esta medida é utilizada para avaliar o relacionamento linear entre duas variáveis contínuas.
* É adequada quando os dados seguem uma distribuição normal e quando a relação entre as variáveis é linear.
* A correlação de Pearson varia de -1 a 1, onde -1 indica uma relação negativa perfeita, 0 indica ausência de correlação e 1 indica uma relação positiva perfeita.
* Por exemplo, você pode usar a correlação de Pearson para investigar a relação entre a altura e o peso das pessoas, pois essas variáveis ​​podem estar linearmente relacionadas.



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.datasets import make_classification, make_regression
```

```python
X, y = make_regression(n_samples=100, n_features=5, n_informative=10)

# Colocar dados em um Dataframe Pandas para melhor manipulação
df = pd.DataFrame(X)
df['target'] = y
df.columns = ['var1', 'var2', 'var3', 'var4', 'var_cat', 'target']
```

```python
# Criando variável Categórica baseada na variável target - propositalmente, para termos forte correlação
df['var_cat'] = pd.cut(df.target,4, labels=[1,2,3,4])
df['var_cat'] = df.var_cat.astype(int)
```

```python
df.corr()
```

```python
df.corr(method='pearson')
```

```python
# Correlação Pearson e Spearman
fig, ax = plt.subplots(figsize=(6,4))
g = sns.heatmap(df.corr(method='pearson'), cmap='coolwarm', annot=True)
g.set_title('Correlação Pearson');
```

# Correlação de Spearman

* Correlação entre variáveis numéricas (lineares ou não) ou entre variáveis qualitativas e quantitativas.
* Correlação de Spearman é para valores lineares ou não-linerares e para variáveis numéricas ordinais (3 > 2 > 1)
* Essa medida é uma alternativa à correlação de Pearson quando os dados não seguem uma distribuição normal ou quando há uma relação não linear entre as variáveis.
* A correlação de Spearman avalia a associação monotônica entre duas variáveis, ou seja, se, quando uma variável aumenta, a outra tende a aumentar ou diminuir consistentemente.
* Ela não assume a linearidade da relação e é mais robusta a outliers.
* A correlação de Spearman é calculada a partir das classificações dos dados, ou seja, ela avalia a relação entre as ordens dos dados, não seus valores absolutos.
*Por exemplo, você pode usar a correlação de Spearman para investigar a relação entre a classificação de desempenho de alunos em dois exames diferentes, onde a relação pode ser monotônica, mas não necessariamente linear.

```python
df.corr(method='spearman')
```

```python
# Correlação Pearson e Spearman
fig, ax = plt.subplots(figsize=(6,4))
g = sns.heatmap(df.corr(method='spearman'), cmap='coolwarm', annot=True)
g.set_title('Correlação Spearman');
```

```python
# Correlação Pearson e Spearman
fig, ax = plt.subplots(1,2,figsize=(20,8))
g = sns.heatmap(df.corr(method='pearson'), cmap='coolwarm', annot=True, ax=ax[0])
g.set_title('Correlação Pearson')
g2 = sns.heatmap(df.corr(method='spearman'), cmap='coolwarm', annot=True, ax=ax[1])
g2.set_title('Correlação Spearman');
```

## Variáveis numéricas:

* Se ambas as variáveis forem numéricas e houver suposição de linearidade e distribuição normal, a correlação de Pearson é geralmente a escolha preferida.
*Se não houver suposição de linearidade ou distribuição normal, ou se os dados forem muito influenciados por outliers, a correlação de Spearman pode ser mais apropriada.

## Variáveis categóricas:

* Para variáveis categóricas ordinais, a correlação de Spearman é frequentemente utilizada, pois ela considera a ordem dos dados, não os seus valores absolutos.
* Para variáveis categóricas nominais, outras medidas de associação, como coeficiente de contingência ou teste qui-quadrado, são mais adequadas, pois tanto Pearson quanto Spearman requerem variáveis numéricas.

# Chi quadrado

* Mede a relação de dependência entre duas variáveis categóricas, verificando como os valores esperados desviam dos valores observados.
* Valores altos de Chi-Quadrado indicam variáveis mais dependentes e melhores para serem usadas no modelo.
* Quando temos um alto valor de Chi-quadrado (nosso p-value será baixo), significa que temos evidência estatística para inferir que os valores observados e esperados não são os mesmos, portanto possuem dependência entre si.

```python
df = sns.load_dataset(name='penguins')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
df1 = df.copy()
df1['species'] = df1.species.map({'Adelie':1, 'Chinstrap':2, 'Gentoo':3})
df1['island'] = df1.island.map({'Torgersen':1, 'Biscoe':2, 'Dream':3})
df1['sex'] = df1.sex.map({'Male':0, 'Female':1})
df1.head()
```

```python
from sklearn.feature_selection import chi2
```

```python
# Split dataset
X = df1.drop(['sex', 'bill_length_mm', 'bill_depth_mm','flipper_length_mm','body_mass_g'], axis=1)
y = df1.sex
```

```python
# Calculando Chi-Squared
chi_scores = chi2(X,y)

# Colocando label nos resultados para posterior plot
scores = pd.Series(chi_scores[0], index=X.columns)
pvalues = pd.Series(chi_scores[1], index=X.columns)

# Results
pd.DataFrame({'Chi2':scores, 'p-Value':pvalues})
```

Como resultado, temos que os **menores Chi2 scores/ maiores p-Values** são os atributos menos dependentes da variável target Sexo.

Portanto, temos que a espécie (species) e a ilha onde habitam (island) são independentes do sexo do animal, não influenciando em uma possível previsão.
Isso faz todo o sentido, dado que em todas as ilhas e em todas as espécies, teremos os dois sexos de pinguins, sendo muito difícil prever essa variável com apenas essa informação.

# Anova

* ANOVA é um método estatístico para verificar se existe diferenças significativas entre as médias de grupos de dados, sendo possível inferir se as variáveis são dependentes uma sobre a outra.
* medir o F-Value que é a relação entre a variância entre grupos dividido pela variância dentro dos grupos.
* Inversamente verdadeiro, maiores valores de F significam menores valores de p — significância estatística.
* A Ho (p-value alto) é de que os grupos tem a mesma variância e, portanto, não tem impacto para treinamento de modelo.
*Halt (p-value baixo) de que existe diferença significativa e o atributo é bom para treinamento de modelo.

```python
# Calculando a estatística com ANOVA
aov = []
for variance in df1.columns[:6]:
  v = pg.anova(data=df1, dv='sex', between=variance, detailed=True).round(3)
  aov.append(v)
```

```python
pd.concat(aov)
```

* SS (Sum of Squares): Esta coluna refere-se à soma dos quadrados dos desvios das médias em relação à média global. É uma medida da variabilidade total dos dados.

* DF (Degrees of Freedom): Graus de liberdade. No contexto da ANOVA, este é o número de categorias menos um. Para o tratamento, é o número de grupos menos um. Para o erro, é o número total de observações menos o número de grupos. Para o total, é o número total de observações menos um.

* MS (Mean Square): É a soma dos quadrados dividida pelos graus de liberdade. Representa a variabilidade média dentro de cada grupo.

* F: A estatística F é a razão entre a variabilidade entre os grupos e a variabilidade dentro dos grupos. É usado para testar se há diferença significativa entre as médias dos grupos.

* p-unc (p-value): O valor p (p-value) é a probabilidade de observar uma estatística de teste igual ou mais extrema do que a observada, sob a hipótese nula de que não há efeito ou diferença. É usado para determinar se os resultados da ANOVA são estatisticamente significativos.

* np2 (Partial Eta-squared): Eta ao quadrado parcial é uma medida de tamanho do efeito na ANOVA. Ele fornece uma estimativa da proporção da variância total do resultado que pode ser atribuída a um determinado fator. Em outras palavras, quantifica a magnitude do efeito do fator independentemente do número de níveis desse fator.

```python
data = pd.concat(aov)
data = data[data.Source!='Within'][['Source','F']]
data
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='Source', y='F', data=data)
plt.show()
```

Verificando o resultado das análises de variância, chegamos ao mesmo resultado. As variáveis mais relevantes para construção de um modelo seriam bill_depth, body_mass_g, Flipper_length e bill_length, sendo as espécies e ilhas onde eles vivem menos relevantes para prever o sexo do animal.
